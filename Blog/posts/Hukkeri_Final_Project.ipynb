## Introduction

---



<p>The dataset title “Top Intelligent People in the World” offers an interesting glimpse into the academic lives and achievements of some of the most intellectually gifted individuals across the globe. This dataset provides detailed records of 5,000 leading minds on various aspects including their names, professions, contributions, and IQ scores. This project aims to ascertain how the educational background of some of the world’s most intelligent people influenced their fields of expertise and the types of achievements they attained while also exploring other motivations.</p>

<p>I was also particularly interested in understanding how we could characterize intelligence in addition to quantitative measures such as IQ, and what different factors could contribute to high intelligence. Such factors could include education, environment, personal interests, and more. This dataset includes individuals from the year 1900 to 2000, which offers us the opportunity to explore how different cultures and time periods have cultivated intelligence. We can further explore how historical and cultural factors influenced intelligence and the achievements brought forth by the recorded individuals. Since the dataset includes both male and female individuals, we can further analyze how historical perspectives coupled with educational opportunities offered affected the intelligence/contributions of women compared to men. The dataset further includes the individuals’ country of origin, which can shed light on the educational resources available in each country and how these resources may have impacted the number of highly intelligent individuals it produced or the types of achievements they attained.</p>

<p>Amongst the above motivations that drive my exploration of this dataset, below are some of the questions that I seek to address:</p>
<ul>
<li>Which fields of expertise do the most intelligent people excel in? Are there certain professions or areas of study that attract high IQ individuals more than others?</li>
<li>How has the recognition and distribution of high intelligence changed over time? Are there historical periods that produced more recognized geniuses?</li>
<li>Is there a correlation between IQ scores and the level of achievement or impact these individuals have had in their respective fields?</li>
<li>What are the demographic patterns amongst the top intelligent people? Are there commonalities in age, gender, nationality, or educational background?</li>
<li>What environmental factors (e.g. socioeconomic status, educational background, cultural influences, etc) are common amongst these individuals? How do these factors influence the development of high intelligence?</li>
</ul>

These findings will primarily be revealed in our preliminary analysis.

```python
achievements_count = df["Achievements"].value_counts()
achievements_count
```

```python
notable_works_count = df["Notable Works"].value_counts()
notable_works_count
```

```python
plt.figure(figsize = (12,6))
plt.bar(notable_works_count.index[:10], notable_works_count[:10])
plt.title("Fig 8. Top 10 Notable Works", fontsize=16)
plt.xlabel("Notable Works", fontsize=11)
plt.ylabel("Number of Individuals", fontsize=11)
plt.xticks(rotation=85, fontsize=12)
plt.show()
```

```python
# define the bin edges for 20-year intervals
bins = range(1900, 2000, 20)
labels = [f"{year}-{year+20}" for year in bins[:-1]]

# 'Period' column based on the bins
df['Period'] = pd.cut(df['Birth Year'], bins=bins, labels=labels, right=False)
top_works = notable_works_count.index[:5]

# store counts of top 5 notable works by period
grouped_counts = pd.DataFrame()
for work in top_works:
    work_counts = df[df['Notable Works'] == work].groupby('Period', observed=False).size()
    grouped_counts[work] = work_counts

fig, ax = plt.subplots(figsize=(16, 8))
width = 0.15
x = range(len(grouped_counts))  # X locations for groups

for i, work in enumerate(top_works):
    ax.bar([p + width * i for p in x], grouped_counts[work], width=width, label=work)

ax.set_xlabel('Birth Year', fontdict={'fontsize': 12})
ax.set_ylabel('Number of Individuals', fontdict={'fontsize': 12})
ax.set_title('Fig 9. Frequency of Top 5 Notable Works Over Time', fontsize=16)
ax.set_xticks([p + width * (len(top_works) - 1) / 2 for p in x])
ax.set_xticklabels(labels, rotation=45, fontsize=12)
ax.legend(title='Notable Works')
plt.show()
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from collections import Counter
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
nltk.download('stopwords')
from string import punctuation
```

```python
notable_works_count
```

```python
all_words = ' '.join(df['Notable Works'])
tokens = word_tokenize(all_words)
filtered_tokens = [word for word in tokens if word.isalpha()]
not_useful = ['discovery', 'equation', 'research', 'paradox', 'theory', 'origin', 'two', 'chief', 'various']
stop_words = list(punctuation) + stopwords.words('english') + not_useful
filtered_tokens = [word for word in filtered_tokens if word.lower() not in stop_words]
word_freq = Counter(filtered_tokens)

# the most common words
most_common_words = word_freq.most_common(15)
most_common_words
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# numerical features used for clustering
features = df[['IQ', 'Birth Year', 'Gender']].copy()
features['Gender'] = features['Gender'].apply(lambda x: 1 if x == 'Male' else 0)

# standardize the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# apply k-means clustering
kmeans = KMeans(n_clusters=5, n_init=10, random_state=42)
features['Cluster'] = kmeans.fit_predict(scaled_features)
features.groupby('Cluster').mean()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(features['IQ'], features['Birth Year'], c=features['Cluster'], cmap='viridis', marker='o')
plt.xlabel('IQ')
plt.ylabel('Birth Year')
plt.title('Clusters Visualization')
plt.colorbar(label='Cluster')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

features_2 = features[['IQ', 'Birth Year', 'Gender']].copy()
target = df['Field of Expertise']

# split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)

# train a RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)

# make predictions
y_pred = clf.predict(X_test)

# evaluate
print(classification_report(y_test, y_pred))
```

```python
education_mapping = {
    "Ph.D. in Physics": 1,
    "Ph.D. in Mathematics": 2,
    "Self-taught": 3,
    "Ph.D. in Chemistry": 4,
    "Ph.D. in Astronomy": 5,
    "Ph.D. (honorary)": 6,
    "University of Cambridge": 7,
    "University of Pisa": 8,
    "Ph.D.": 9
}

df['Education_Numeric'] = df['Education'].map(education_mapping)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import math

# select features & target variable
features = df[['Birth Year', 'Gender', 'Education_Numeric']].copy()
features['Gender'] = features['Gender'].apply(lambda x: 1 if x == 'Male' else 0)
target = df['IQ']

# standardize features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# split data into train & test sets
X_train, X_test, y_train, y_test = train_test_split(features_scaled, target, test_size=0.3, random_state=42)

# train LinearRegression model
reg = LinearRegression()
reg.fit(X_train, y_train)

# prediction
y_pred = reg.predict(X_test)

# evaluation
mse = mean_squared_error(y_test, y_pred)
print(f'Root Mean Squared Error: {math.sqrt(mse)}')
```

## Conclusion

---



A lot of the preliminary analysis plots showed fairly uniform distributions across different measures, but this might simply indicate that this is a highly curated dataset. However, this does not imply that our results are without value. Figure 2 indicates that certain time periods didn’t produce more intelligent people and that the distribution of people recorded remained fairly uniform over time. However, it’s important to note that this isn’t a definitive indication as 5,000 individuals is simply a small sample of a much larger population. Figure 4 indicates that the United Kingdom has the most number of intelligent people. It also shows that besides the United States, other individuals in the dataset originated from other European countries, which might be due to the fact that European countries had the best research facilities in the 20th century that allowed for STEM advancements, or it might indicate a clear bias in the dataset.<br>
<br>
Deeper analysis into the achievements, notable works, education, and influence of these individuals proved to be interesting. The quantum revolution of the 20th century was evidently the result of many of these brilliant minds. Many experimental and theoretical physicists began to test the limits of manipulating matter on the atomic scale, bringing forth the emergence of new technologies, all dependent on quantum physics (Prawer, 2018). Technologies such as semiconductors and computers, lasers and communications, MRIs, PET imaging, etcetera all arose thanks to quantum computing and quantum mechanics. Figure 7 supports this as most of the individuals in this dataset had achievements in “Quantum Mechanics”, and further in “Quantum Electrodynamics”, and “Quantum Theory.” Additionally, Figure 8 presents that many notable works stemmed from Quantum Theory, and Figure 9 furthers the sentiment that the frequency of notable works in quantum theory increased over time. This is evidenced by the fact that as more bright individuals were born, the more quantum-related notable works were produced.<br>
<br>
Textual analysis of notable works shed insights into what field was prominent in the 20th century. The words “Dirac” and “Quantum” appeared the most. The latter we have covered, but the former is part of the notable work, “Dirac Equation”. Discovered by the theoretical physicist Paul Dirac, who set work to look for “pretty mathematics,” the equation reconciled “two seemingly contradictory characteristics of the world” revealed in experiments in the first quarter of the twentieth century: the ability of atoms to behave as both localized particles and as spread-out waves (Chown, 2023).  Furthermore, this historical period showed a high frequency in attaining Ph.D.’s in Physics and Mathematics (Figure 10). Figure 11 simply promotes these findings as it shows the top 10 areas of influence heavily include those in quantum and modern physics. Additionally, it can be inferred that the large number of Nobel Prizes attained were likely in physics and mathematics, which were prominent fields in that time period (Figure 12).<br>
<br>
Moreover, “Cosmos” was the third most common word in notable works. In addition to physics, discoveries in astronomy revolutionized our view of the cosmos. Einstein’s general theory of relativity, though sidelined from the mainstream of physics during the early part of the century, proved vital to astronomers' new theories about our universe. The universe was not limited to the Milky Way and was bigger than anyone thought (Rees, 2020). What’s interesting, and also quite obvious, is that both physics and astronomy went hand in hand to help these brilliant individuals make the breakthroughs they made. Similarly, mathematics and physics both also played a part, underscoring the interdisciplinary nature of many of the advancements brought about by them. This dataset was very eye opening about the impact that many men and women have had in the way society thinks about the world.<br>
<br>
I applied a clustering algorithm on the numerical features of the dataset to see if I could identify any relationships or patterns that may not have been immediately obvious in our dataset. I hoped to group individuals based on their IQ, birth year, and gender, which might have given us insights into how these features correlate and form distinct groups. Though the clusters visualization shows the different clusters formed, we can see they aren’t densely clustered in any one area, indicating that it isn’t capturing any stark patterns in the data. I additionally trained a random forest classifier to predict the field of expertise an individual in the dataset would have based on IQ, birth year, and gender. Though most of the fields had very low precision, recall, and f1-scores, the model had higher precision (0.43) and recall (0.55) for the “Physics” class, indicating that it performed better for this category compared to others. This result is unsurprising considering what has been discussed thus far. I also performed linear regression on the dataset to try to predict the IQ of individuals based on their birth year, gender, and education type; however, the resulting root mean squared error of 20.5 indicates what the average root squared difference between the predicted and the actual IQ values is. Since the IQ values range from 130 to 200, the average error might be substantial, suggesting that the model’s predictions were not very accurate.<br>
<br>
While we clearly saw many notable accomplishments in STEM fields, also evident by everything we know as a society today, there is so much more that can be done with such a dataset. We can derive insights to help policymakers and educators develop educational programs to maximize its contributions to society by exploring beyond the dataset.
